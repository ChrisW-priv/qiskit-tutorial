```python
from qiskit import QuantumCircuit
```

```python
qc=QuantumCircuit(2,1)
qc.i(0)
qc.x(1)
qc.h(0)
qc.h(1)
qc.barrier(range(2))
qc.cx(0,1)
qc.barrier(range(2))
qc.h(0)
qc.h(1)
qc.barrier(range(2))
qc.measure(0,0)
qc.draw(output='mpl')
```

```python
from qiskit_aer import AerSimulator as Aer
from qiskit import transpile
backend=Aer()
qc_compiled = transpile(qc, backend)
job_sim = backend.run(qc_compiled, shots=1024)
result_sim = job_sim.result()
```

```python
counts = result_sim.get_counts(qc_compiled)
print(counts)
```

```python
qc1=QuantumCircuit(2,2)
qc1.i(0)
qc1.x(1)
qc1.h(0)
qc1.h(1)
qc1.barrier(range(2))
qc1.draw('mpl')
```

```python
qc2=QuantumCircuit(2,2)
qc2.cx(0,1)
qc2.barrier(range(2))
qc2.draw('mpl')
```

```python
qc3=QuantumCircuit(2,2)
qc3.h(0)
qc3.h(1)
qc3.barrier(range(2))
qc3.measure(0,0)
qc3.draw('mpl')
```

```python
qc = qc1.compose(qc2).compose(qc3)
qc.draw('mpl')
```

```python
qc2=QuantumCircuit(2,6)
qc2.measure(0,0)
qc2.measure(1,1)
qc2.barrier(range(2))
qc2.x(1)
qc2.id(0)
qc2.measure(0,2)
qc2.measure(1,3)
qc2.barrier(range(2))
qc2.cx(0,1)
qc2.barrier(range(2))
qc2.measure(0,4)
qc2.measure(1,5)
qc2.draw('mpl')
```
